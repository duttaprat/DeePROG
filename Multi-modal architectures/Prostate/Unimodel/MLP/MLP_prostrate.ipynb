```python
X_train,X_test,Y_train,Y_test=train_test_split(augmented_X,augmented_Y,test_size=0.20,random_state=2)
print(X_test.shape)
```

```python
class Net(torch.nn.Module):
    def __init__(self, n_feature,out_features):
        super(Net, self).__init__()
        self.hidden1 = torch.nn.Linear(n_feature,900)
        self.hidden2 = torch.nn.Linear(900,600)
        self.hidden3 = torch.nn.Linear(600,400)
        self.hidden4 = torch.nn.Linear(400,100)
        self.hidden5 =torch.nn.Linear (100,10)
        self.predict = torch.nn.Linear(10,out_features)
        self.bn1 = nn.BatchNorm1d(num_features=900)
        self.bn2 = nn.BatchNorm1d(num_features=600)
        self.bn3 = nn.BatchNorm1d(num_features=400)
        self.bn4 = nn.BatchNorm1d(num_features=100)
        self.bn5 = nn.BatchNorm1d(num_features=10)
    def forward(self, x):
        x = F.relu(self.hidden1(x))
#         x=self.bn1(x)
        x = F.relu(self.hidden2(x))
#         x=self.bn2(x)
        x =F.relu(self.hidden3(x))
        x=self.bn3(x)
        x =F.relu(self.hidden4(x))
        x=self.bn4(x)
        x =F.relu(self.hidden5(x))
#         x=self.bn5(x)
        
        x=F.relu(self.predict(x))
        return x
def accurate():
    total = X_test.shape[0]
    x_test=torch.tensor(X_test,dtype=torch.float)
    y_test=torch.tensor(Y_test,dtype=torch.long)
    outputs = MP(x_test)
    _, predicted = torch.max(outputs.data, 1)
    correct = (predicted.cpu() == y_test).sum()
    confusion_matrix = torch.zeros(2, 2)
    print('Accuracy of the network on the test data: %d %%' % (100 * correct / total))
    for t, p in zip(y_test.view(-1), predicted.view(-1)):
        confusion_matrix[t.long(), p.long()] += 1
    print(confusion_matrix)
```

```python
MP=Net(104,2)

inp=torch.tensor(X_test,dtype=torch.float)

criterion = nn.CrossEntropyLoss()
target=torch.tensor(np.zeros((4)),dtype=torch.long)
#output = criterion(mid, target)

optimizer = torch.optim.Adam(MP.parameters(), lr = 0.01)

epoch=100
batch=10

sizez=(np.array(X_train)).shape[0]
print (sizez)


```

```python

epoch=10000
def trainz():
    no_batch=(sizez//batch)
    for i in range (epoch):
        start=0
        NET_LOSS=0
        for j in range(no_batch+1):

            stop=start+batch
            if(stop<=sizez):
                  stop=stop
            else:
                  stop=sizez
        if start>=sizez:
              break

        x_batch=torch.tensor(X_train[start:stop],dtype=torch.float)
        y_batch=torch.tensor(Y_train[start:stop],dtype=torch.long)

        out_batch=MP(x_batch)
        LOSS_BATCH=criterion(out_batch,y_batch)
        optimizer.zero_grad()
        LOSS_BATCH.backward()
        optimizer.step()
        NET_LOSS=NET_LOSS+LOSS_BATCH
#       print(str(LOSS_BATCH)+"___"+str(j))
        start=start+batch
        if start>=sizez or start>stop:
              break
        accurate()
        print(NET_LOSS)

trainz()
```
